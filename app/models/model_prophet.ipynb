# MediFlow AI — Prophet Forecaster Model Training

This notebook contains the local training and evaluation pipeline for the **Prophet Bed Occupancy Forecaster**.

```python
import os
import joblib
import numpy as np
import pandas as pd
from prophet import Prophet

print("Prophet forecaster initialized.")
```

```python
# Load processed forecast dataset
PROCESSED_FORECAST_CSV = "../../data/processed/hourly_counts.csv"
if not os.path.exists(PROCESSED_FORECAST_CSV):
    PROCESSED_FORECAST_CSV = "data/processed/hourly_counts.csv"

df = pd.read_csv(PROCESSED_FORECAST_CSV)
print("Loaded time series shape:", df.shape)
print(df.head())
```

```python
# Train Prophet model with daily and weekly seasonality
model = Prophet(
    growth="linear",
    yearly_seasonality=False,
    weekly_seasonality=True,
    daily_seasonality=True
)
model.fit(df)
print("Model fit complete.")
```

```python
# Save Prophet model
joblib.dump(model, "model_prophet.pkl")
print("Prophet model successfully saved to 'model_prophet.pkl'.")
```
