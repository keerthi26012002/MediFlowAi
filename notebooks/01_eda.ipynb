# MediFlow AI — Exploratory Data Analysis

This notebook covers:
- Dataset shape, column names, and types verification.
- Mapping raw attributes to our logical domain.
- Null values and duplicates analysis.
- Wait-times and admissions correlations.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_theme(style='darkgrid')
plt.rcParams['figure.figsize'] = (10, 6)

# Load dataset
df = pd.read_csv('../datasets/Hospital ER_Data.csv')
print(f'Shape of raw dataset: {df.shape}')
print('\nColumns list:')
print(df.columns.tolist())
print('\nData Types:')
print(df.dtypes)
```

```python
print('First 5 records:')
df.head()
```

```python
print('Missing values per column:')
print(df.isnull().sum())
```

```python
# Wait time distribution
sns.histplot(df['Patient Waittime'], kde=True, bins=30, color='blue')
plt.title('Patient Wait Time Distribution')
plt.xlabel('Wait Time (minutes)')
plt.ylabel('Patient Count')
plt.show()
```

```python
# Admission rate by Department
df['Department Referral'] = df['Department Referral'].fillna('Self-Referral')
sns.countplot(data=df, x='Department Referral', hue='Patient Admission Flag')
plt.title('Patient Admission Flag by Department Referral')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
